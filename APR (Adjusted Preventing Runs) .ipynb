```python
import pandas as pd
```

```python
excel_file_path = 'C:/Users/ASUS/OneDrive/Documents/實習/Rebas/CPBL_2023.xlsx'
```

```python
sheet_name3 = 'AWAY-BOX-投球'  # 請替換成你要載入的工作表名稱
sheet_name4 = 'HOME-BOX-投球'
away1 = pd.read_excel(excel_file_path, sheet_name=sheet_name3)
home1 = pd.read_excel(excel_file_path, sheet_name=sheet_name4)
```

```python
away1.rename(columns={'客場': '球隊'}, inplace=True)
home1.rename(columns={'主場': '球隊'}, inplace=True)
```

```python
merged_stat = pd.concat([away1, home1], ignore_index=True)

#merged_stat = merged_stat.drop(["比賽編號","比賽日期","客場","主場","背號"], axis = 1)
merged_stat
```

```python
park_factors = {
    "臺北市立天母棒球場": 0.974586135,
    "臺南市立棒球場": 1.049876521,
    "新北市立新莊棒球場": 0.982791461,
    "樂天桃園棒球場": 1.096458186,
    "臺中市洲際棒球場":0.909723723,
    "嘉義市立棒球場":0.938006629,
    "斗六棒球場":1.054394354,
    "澄清湖棒球場":0.932267771,
    "花蓮縣立德興棒球場":1.012278069,
    "台東棒球村第一棒球場":0.726667302,
    "新竹市立棒球場":0.298210736
}

merged_stat["PF"] = merged_stat["球場"].map(park_factors)

merged_stat
```

```python
agg_dict = {
    "出局數": "sum",
    "面對打者": "sum",
    "安打": "sum",
    "全壘打": "sum",
    "保送": "sum",
    "故意四壞": "sum",
    "觸身": "sum",
    "三振": "sum",
    "暴投": "sum",
    "投手犯規": "sum",
    "失分": "sum",
    "責失": "sum",
    "PF":"mean"
}
pitcher_total = merged_stat.groupby(["名稱",'球隊']).agg(agg_dict).reset_index()
pitcher_total
```

```python
pitcher_total["IP"] = pitcher_total["出局數"] / 3
pitcher_total["ERA"] = (pitcher_total["責失"] / pitcher_total["IP"]) * 9
pitcher_total["FIP"] = (13 *pitcher_total["全壘打"] + 3 * pitcher_total["保送"] + 2 * pitcher_total["觸身"] + 3 * pitcher_total["暴投"] - 2 * pitcher_total["三振"]) / pitcher_total["IP"] + 3.2
pitcher_total["K%"] = (pitcher_total["三振"] / pitcher_total["面對打者"]) * 100
pitcher_total["BB%"] = (pitcher_total["保送"] / pitcher_total["面對打者"]) * 100
```

```python
lg_ERA = (pitcher_total["責失"].sum() / (pitcher_total["出局數"].sum()/3).sum()) * 9
lg_FIP =(13 *pitcher_total["全壘打"].sum() + 3 * pitcher_total["保送"].sum() + 2 * pitcher_total["觸身"].sum() + 3 * pitcher_total["暴投"].sum() - 2 * pitcher_total["三振"].sum()) / (pitcher_total["出局數"].sum()/3) + 3.2
lg_RA = pitcher_total["失分"].sum()
lg_inn = pitcher_total["出局數"].sum()/3
```

```python
print(lg_ERA)
print(lg_FIP)
print(lg_RA)
print(lg_inn)
```

```python
pitcher_total["ERA+"] = (lg_ERA / pitcher_total["ERA"]) * 100
pitcher_total["FIP+"] = (lg_FIP / pitcher_total["FIP"]) * 100
```

## APR (Adjusted Preventing Runs)  計算

### 公式:APR = IP / 9 * (LgERA - ERA)						　
#### 修正失分=投球局數/9*(聯盟平均ERA-ERA)

### 當ERA已經經過球場修正後，這個指標可以測量出一個投手與一個聯盟平均等級的投手在一個「中立」的球場投球時，在同樣的投球局數下，可以少失多少(自責)分。
### 數字越高代表可省下的分數越多

#### *說明摘自https://ej-baseball.com/stastics/Sabermetric_Pitching.htm

```python
pitcher_total['APR']=pitcher_total['IP']/9*(lg_ERA - pitcher_total['ERA'])
```

## APR (Adjusted Preventing Runs) 計算2.0 

### 將Park Factor計算進去
### 公式 APR = L * IP - R / pf(P)

#### L: Average number of runs per inning pitched in the league 
#### IP: Innings Pitched
#### R: Runs Allowed
#### pf(P): park factor for the player's home park P.

##### 參考網址:https://legacy.baseballprospectus.com/glossary/index.php?mode=viewstat&stat=233

```python
pitcher_total['APR(PF)'] = (lg_RA/lg_inn)*pitcher_total['IP']-pitcher_total['失分']/pitcher_total['PF']
```
